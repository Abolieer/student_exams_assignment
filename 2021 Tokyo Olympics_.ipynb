```python
%pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt 
import os 
import pandas as pd 
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
print("The statistics of 2021 Tokyo Olympics")
```

<font size="5">Gold Medal</font>

```python
print("This infographic is a summary of the medal in the 2021 Tokyo Olympics. It shows Top 15 countries who get at least one gold medal at the Tokyo Olympics.")
```

```python
Gold_results = pd.read_csv('2021.csv', encoding='latin-1')
sns.set(rc={'figure.figsize':(10,7)})
sns.barplot(x = 'Gold Medal', y = 'Country',data = Gold_results, ci = None,order=Gold_results.sort_values('Gold Medal',ascending = False).Country, color ="gold" )
plt.ylim(-0.5, 14.5)
```
